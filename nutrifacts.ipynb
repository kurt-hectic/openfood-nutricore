```python
fileurl = 'http://fr.openfoodfacts.org/data/fr.openfoodfacts.org.products.csv'
filename_foodfacts = 'fr.openfoodfacts.org.products.csv'

if not os.path.isfile(filename_foodfacts):
    urllib.request.urlretrieve(fileurl,filename_foodfacts)
```

```python
open_food_repo_data=json.load( open('dInfo.json') )

print(" number of products in the JSON file {} ".format( len(open_food_repo_data.keys()) ) )

df_open_food =  pd.DataFrame.from_dict(  open_food_repo_data )

df_open_food = df_open_food.transpose()
```

```python
df_foodfacts =df_foodfacts.set_index('code')

df_join = pd.merge(left=df_foodfacts , right=df_open_food )
```
